# Exploratory Analysis and Visualization

### Highest Total Profit earned among different Item Types

It'd be interesting to see the comparison between the total profit earned for each an every itemtype in this dataset to get an idea of the total profit earned.

```python
q = data_5000.copy()
q_a= q.groupby('Item Type')[['Total Profit']].sum()# as we need the individual profit of each item we group item type then sum on profit. .
q_a.reset_index(inplace=True)# we use reset index to the index column as 1,2,3...
q_a.sort_values(by='Total Profit',inplace = True)# sort values is used to sort in the data frame according to column.
```

```python
q_a
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))
plt.bar(q_a['Item Type'], q_a['Total Profit'], edgecolor='red');
```

#### So there are roughly 12 types of items and profits are varying accordingly . This make sense since as per the demand ,rate of the item gets changed.Here the highest profit shows for cosmetics.

## How many order's are from each country 

#### Now, we will explore the number of orders has been done for each of the diferent countries.

```python
q_3 = data_5000.groupby(['Country'])[['Order ID']].count()
q_3.reset_index(inplace = True)
```

```python
q_3
```

```python
plt.figure(figsize=(15,30))
plt.xlabel('Orders')
plt.ylabel("Countries")
plt.title(' NUmber of Orders')
plt.plot(q_3['Order ID'],q_3['Country'],marker='*', color='g', ls='--', lw=3, ms=10, alpha=.9);
```

#### From the above graph we can clearly observe that the line is straight up inclination i.e maximum value 40 order's for Andorra , San Marino and minimum value 13 is Afghanistan.
#### Overall, there are much more orders are done across different countries.

# How many order's are from each Region.

#### Now we will take a look at the number of orders has done throughout the each region.

```python
s = data_5000.groupby(['Region'])[['Order ID']].count()
s.reset_index(inplace = True)
s.sort_values(by='Order ID',inplace = True)
```

```python
s
```

```python
plt.figure(figsize=(10,8))
plt.title("Regions Vs Number of Order's")
sns.barplot(x = 'Order ID',y = 'Region' ,data = s)



for index,data1 in enumerate(s['Order ID']): #enumerate as we use for 2values in for loop here its triversing in "Order ID"
    plt.text(x=data1 , y =index , s=(f"{data1}") , fontdict=dict(fontsize=20));
```

#### Based on the above bar graph, we can see that the different type of the regions has different number of orders the maximum orders are done in the Europe region and miminum in North America

## Year wise Total Profit vs Every item (every year)

#### Now, we will take a look at the Year wise total profit for every item.

```python
data_5000
```

```python
o = data_5000.groupby(['year','Item Type'])[['Total Profit']].sum()
```

```python
o
```

```python
Year_2010 = data_5000[data_5000['year']== 2010]
Year_2011 = data_5000[data_5000['year']== 2011]
Year_2012 = data_5000[data_5000['year']== 2012]
Year_2013 = data_5000[data_5000['year']== 2013]
Year_2014 = data_5000[data_5000['year']== 2014]
Year_2015 = data_5000[data_5000['year']== 2015]
Year_2016 = data_5000[data_5000['year']== 2016]
Year_2017 = data_5000[data_5000['year']== 2017]
```

#### In the above dataframe there are 8 different years and it has different profit for different year along with item type ,From this we can analyse the how much profit is gained for particular year in particular year.

```python
# creating new dataframes for Item Types every year along with the total profit.
Q_5 = Year_2010.groupby('Item Type')[['Total Profit']].sum()
Q_5.reset_index(inplace=True)


Q_6 = Year_2011.groupby('Item Type')[['Total Profit']].sum()
Q_6.reset_index(inplace=True)


Q_7 = Year_2012.groupby('Item Type')[['Total Profit']].sum()
Q_7.reset_index(inplace=True)


Q_8 = Year_2013.groupby('Item Type')[['Total Profit']].sum()
Q_8.reset_index(inplace=True)


Q_9 = Year_2014.groupby('Item Type')[['Total Profit']].sum()
Q_9.reset_index(inplace=True)


Q_10 = Year_2015.groupby('Item Type')[['Total Profit']].sum()
Q_10.reset_index(inplace=True)


Q_11 = Year_2016.groupby('Item Type')[['Total Profit']].sum()
Q_11.reset_index(inplace=True)


Q_12 = Year_2017.groupby('Item Type')[['Total Profit']].sum()
Q_12.reset_index(inplace=True)
```

```python
fig, axes = plt.subplots(2,4, figsize=(60,30))

axes[0,0].set_title('Year -- 2010',fontsize=30)#--------->1
axes[0,0].set_xlabel('Item Type', fontsize = 30)
axes[0,0].set_ylabel('Total Profit',fontsize = 30)
sns.barplot(x=Q_5['Item Type'], y=Q_5['Total Profit'], ax=axes[0,0])


for index,data in enumerate(Q_5['Total Profit']):
    axes[0,0].text(x=index , y =data , s=int(f"{data}") , fontdict=dict(fontsize=20))


axes[0,1].set_title('Year -- 2011',fontsize=30)#------->2
axes[0,1].set_xlabel('Item Type', fontsize = 30)
axes[0,1].set_ylabel('Total Profit',fontsize = 30)
sns.barplot(x=Q_6['Item Type'], y=Q_6['Total Profit'],palette='bright',ax=axes[0,1])


for index,data in enumerate(Q_6['Total Profit']):
    axes[0,1].text(x=index , y =data , s=int(f"{data}") , fontdict=dict(fontsize=20))

axes[0,2].set_title('Year -- 2012',fontsize=30)#------->3
axes[0,2].set_xlabel('Item Type', fontsize = 30)
axes[0,2].set_ylabel('Total Profit',fontsize = 30)
sns.lineplot(x=Q_7['Item Type'], y=Q_7['Total Profit'],ax=axes[0,2])


for index,data in enumerate(Q_7['Total Profit']):
    axes[0,2].text(x=index , y =data , s=int(f"{data}") , fontdict=dict(fontsize=20))
    


axes[0,3].set_title('Year -- 2013',fontsize=30)#--------->4
axes[0,3].set_xlabel('Item Type', fontsize = 30)
axes[0,3].set_ylabel('Total Profit',fontsize = 30)
mylabels = Q_8['Item Type']
origin = Q_8['Total Profit']
axes[0,3].pie(origin, labels = mylabels,autopct='%1.2f%%');







axes[1,0].set_title('Year -- 2015',fontsize=30)#--------->5
axes[1,0].set_xlabel('Item Type', fontsize = 30)
axes[1,0].set_ylabel('Total Profit',fontsize = 30)
sns.barplot(x=Q_9['Item Type'], y=Q_9['Total Profit'], ax=axes[1,0])


for index,data in enumerate(Q_9['Total Profit']):
    axes[1,0].text(x=index , y =data , s=int(f"{data}") , fontdict=dict(fontsize=20))


axes[1,1].set_title('Year -- 2016',fontsize=30)#------->6
axes[1,1].set_xlabel('Item Type', fontsize = 30)
axes[1,1].set_ylabel('Total Profit',fontsize = 30)
sns.barplot(x=Q_10['Item Type'], y=Q_10['Total Profit'],palette='bright',ax=axes[1,1])


for index,data in enumerate(Q_6['Total Profit']):
    axes[1,1].text(x=index , y =data , s=int(f"{data}") , fontdict=dict(fontsize=20))

axes[1,2].set_title('Year -- 2017',fontsize=30)#------->7
axes[1,2].set_xlabel('Item Type', fontsize = 30)
axes[1,2].set_ylabel('Total Profit',fontsize = 30)
sns.lineplot(x=Q_11['Item Type'], y=Q_11['Total Profit'],ax=axes[1,2])


for index,data in enumerate(Q_7['Total Profit']):
    axes[1,2].text(x=index , y =data , s=int(f"{data}") , fontdict=dict(fontsize=20))
    


axes[1,3].set_title('Year -- 2018',fontsize=30)#--------->8
axes[1,3].set_xlabel('Item Type', fontsize = 30)
axes[1,3].set_ylabel('Total Profit',fontsize = 30)
mylabels = Q_12['Item Type']
origin = Q_12['Total Profit']
axes[1,3].pie(origin, labels = mylabels,autopct='%1.2f%%');
```

#### Now,we will explore the year wise sales of the itemtype.typically the demand gets vary accordingly then the changes in the profits gets occured,there will be rapid high profit in one year and may decrease in another year initially in the year of 2010 cosmetics has more demand and profit, but later in the year of 2018 cosmetics profits are reduced and household profits are more.
